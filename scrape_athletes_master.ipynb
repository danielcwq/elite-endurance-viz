```python
import re

def preprocess_json_string(json_str):
    # Remove any leading/trailing whitespace
    json_str = json_str.strip()

    # Remove control characters outside of string values
    json_str = re.sub(r'[\n\r\t]', '', json_str)

    # Function to escape problematic characters within string values
    def escape_and_clean_string(match):
        string_value = match.group(0)
        # Remove the surrounding quotes
        inner_value = string_value[1:-1]

        # Remove any <a>...</a> tags
        inner_value = re.sub(r'<a\b[^>]*>(.*?)<\/a>', '', inner_value, flags=re.IGNORECASE)

        # Escape backslashes that are not part of an escape sequence
        inner_value = re.sub(r'\\(?!["\\/bfnrtu])', r'\\\\', inner_value)
        # Escape double quotes
        inner_value = inner_value.replace('"', '\\"')
        # Escape unescaped newlines and tabs
        inner_value = inner_value.replace('\n', '\\n').replace('\t', '\\t')
        return f'"{inner_value}"'

    # Apply the regex to escape characters and remove <a> tags within string values
    json_str = re.sub(
        r'(?<=":)"(.*?)(?<!\\)"',  # Match string values after a colon
        escape_and_clean_string,
        json_str
    )

    # Remove extra commas before closing braces/brackets
    json_str = re.sub(r',(\s*[}\]])', r'\1', json_str)

    return json_str
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_full_event_results(base_url, start_page, discipline, cutoff_score=1100):
    page = start_page
    results = []
    stop_collecting = False

    while True:
        url = base_url.replace("&page=1", f"&page={page}")
        response = requests.get(url)
        data = response.text
        soup = BeautifulSoup(data, 'html.parser')
        table = soup.find('table')

        if not table:
            break  # No table found on the page, exit the loop

        rows = table.find_all('tr')[1:]  # Skip the header row

        if not rows:
            break  # No more data rows, exit the loop

        for row in rows:
            cols = row.find_all('td')
            if len(cols) > 1:
                mark = cols[1].text.strip()
                competitor = cols[3].text.strip()
                nationality = cols[5].text.strip()
                location = cols[-3].text.strip()
                date = cols[-2].text.strip()
                result_score = cols[-1].text.strip()

                try:
                    result_score_num = int(result_score)
                except ValueError:
                    result_score_num = None

                if result_score_num is not None and result_score_num < cutoff_score:
                    stop_collecting = True
                    break
                results.append({
                    'Mark': mark,
                    'Competitor': competitor,
                    'Nat': nationality,
                    'Location': location,
                    'Date': date,
                    'Results Score': result_score,
                    'Discipline': discipline  # Include the discipline here
                })

        if stop_collecting:
            break  # Exit the outer loop

        page += 1  # Move to the next page

    df = pd.DataFrame(results)
    return df

def collect_multiple_events(event_list):
    dfs = []
    for event in event_list:
        base_url = event['base_url']
        discipline = event['discipline']
        df = get_full_event_results(base_url, start_page=1, discipline=discipline)
        dfs.append(df)
    combined_df = pd.concat(dfs, ignore_index=True)
    return combined_df
```

```python
def search_athlete(athlete_name):
    try:
        # Step 1: Open the search and input athlete name
        search_button = driver.find_element(By.ID, "open-global-search-button")
        search_button.click()
        search_field = driver.find_element(By.ID, "global-search-field")
        search_field.clear()
        search_field.send_keys(athlete_name)

        # Step 2: Perform the search
        send_button = driver.find_element(By.ID, "global-search-button")
        send_button.click()

        # Step 3: Wait for unique IDs (like 'follow_with_approval_button') to appear to ensure search is completed
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "follow_with_approval_button"))
        )

        # Step 4: Check if the search returned "no account" text (i.e., <p class="text-callout">)
        if driver.find_elements(By.CLASS_NAME, "text-callout"):
            return 'NA', 'NA', 'NA', 'NA'

        # Step 5: Locate the first athlete row and extract metadata
        athlete_row = driver.find_element(By.XPATH, "//ul[@class='athlete-search striped container-fluid']//li[@class='row']")

        athlete_detail = athlete_row.find_element(By.XPATH, ".//div[@class='athlete-details']")
        athlete_link = athlete_detail.find_element(By.XPATH, ".//a[@class='athlete-name-link']")
        athlete_id = athlete_link.get_attribute("data-athlete-id")

        # Step 6: Check the activities section for runs and bike rides
        activities_div = athlete_row.find_element(By.CLASS_NAME, "spans2")

        # Step 7: Check if there are no activities logged
        if activities_div.find_elements(By.XPATH, ".//p[contains(text(), 'None')]"):
            runs = 0
            bike_rides = 0
        else:
            # Extract the number of runs (default to 0 if not found)
            runs_elements = activities_div.find_elements(By.XPATH, ".//p[contains(text(), 'Runs')]")
            if runs_elements:
                runs_text = runs_elements[0].text
                runs = int(runs_text.split()[0])  # Extract the number from the text "26 Runs"
            else:
                runs = 0  # No runs logged

            # Extract the number of bike rides (default to 0 if not found)
            bike_rides_elements = activities_div.find_elements(By.XPATH, ".//p[contains(text(), 'Bike Ride')]")
            if bike_rides_elements:
                bike_rides_text = bike_rides_elements[0].text
                bike_rides = int(bike_rides_text.split()[0])  # Extract the number from the text "1 Bike Ride"
            else:
                bike_rides = 0  # No bike rides logged

        # Step 8: Check if the athlete profile is private or public
        follow_action_div = athlete_detail.find_element(By.CLASS_NAME, "follow-action")
        requires_approval = follow_action_div.get_attribute("data-requires-approval")
        is_private = 'Private' if requires_approval == 'true' else 'Public'

        return athlete_id, runs, bike_rides, is_private

    except Exception as e:
        # Return NA if something went wrong (athlete not found or no data)
        print(f"Error searching for {athlete_name}: {e}")
        return 'NA', 'NA', 'NA', 'NA'
```

```python
def augment_dataframe_with_athlete_data(combined_df):
    # Get unique athlete names from the 'Competitor' column
    athlete_names = combined_df['Competitor'].unique()

    # Initialize a list to store athlete data
    athlete_data_list = []

    # Loop over each unique athlete name
    for name in athlete_names:
        print(f"Searching for athlete: {name}")
        athlete_id, runs, bike_rides, is_private = search_athlete(name)
        athlete_data_list.append({
            'Competitor': name,
            'Athlete ID': athlete_id,
            'Number of Runs': runs,
            'Number of Bike Rides': bike_rides,
            'Profile Visibility': is_private
        })
        time.sleep(0.5)  # Optional: sleep between searches to avoid rate limits

    # Create a DataFrame from athlete_data_list
    athlete_data_df = pd.DataFrame(athlete_data_list)

    # Merge the athlete data back into the combined_df on 'Competitor'
    combined_df = combined_df.merge(athlete_data_df, on='Competitor', how='left')

    return combined_df
```

```python
def reprocess_athlete_data(combined_df, start_index, end_index):
    # Subset the DataFrame to process only the specified rows
    df_to_process = combined_df.iloc[start_index:end_index].copy()

    # Ensure the 'Competitor' column is present
    if 'Competitor' not in df_to_process.columns:
        raise KeyError("Column 'Competitor' not found in the DataFrame.")

    # Loop over each athlete in the specified rows
    for index, row in df_to_process.iterrows():
        name = row['Competitor']
        print(f"Reprocessing athlete: {name}")
        athlete_id, runs, bike_rides, is_private = search_athlete(name)
        # Update the df_to_process DataFrame directly
        df_to_process.at[index, 'Athlete ID'] = athlete_id
        df_to_process.at[index, 'Number of Runs'] = runs
        df_to_process.at[index, 'Number of Bike Rides'] = bike_rides
        df_to_process.at[index, 'Profile Visibility'] = is_private
        time.sleep(0.5)  # Optional: sleep between searches to avoid rate limits

    # Update the original combined_df with the new data
    combined_df.update(df_to_process)

    return combined_df
```

```python
def preprocess_json_string(json_str):
    # Remove any leading/trailing whitespace
    json_str = json_str.strip()

    # Remove control characters outside of string values
    json_str = re.sub(r'[\n\r\t]', '', json_str)

    # Function to escape problematic characters within string values
    def escape_and_clean_string(match):
        string_value = match.group(0)
        # Remove the surrounding quotes
        inner_value = string_value[1:-1]

        # Remove any <a>...</a> tags
        inner_value = re.sub(r'<a\b[^>]*>(.*?)<\/a>', '', inner_value, flags=re.IGNORECASE)

        # Escape backslashes that are not part of an escape sequence
        inner_value = re.sub(r'\\(?!["\\/bfnrtu])', r'\\\\', inner_value)
        # Escape double quotes
        inner_value = inner_value.replace('"', '\\"')
        # Escape unescaped newlines and tabs
        inner_value = inner_value.replace('\n', '\\n').replace('\t', '\\t')
        return f'"{inner_value}"'

    # Apply the regex to escape characters and remove <a> tags within string values
    json_str = re.sub(
        r'(?<=":)"(.*?)(?<!\\)"',  # Match string values after a colon
        escape_and_clean_string,
        json_str
    )

    # Remove extra commas before closing braces/brackets
    json_str = re.sub(r',(\s*[}\]])', r'\1', json_str)

    return json_str

def clean_html(value):
    if value:
        soup = BeautifulSoup(value, 'html.parser')
        return soup.get_text(separator=' ', strip=True)
    return value

def extract_numeric(value):
    if value:
        value = clean_html(value)
        match = re.search(r'[\d\.]+', value)
        if match:
            return float(match.group())
    return None
```

```python
def consolidate_weekly_data(df, start_week=1, end_week=40):
    # Lists to hold weekly data
    athlete_ids = []
    names = []
    week_numbers = []
    date_ranges = []
    distances = []
    times_list = []
    elevations = []

    # List to hold individual activity data
    activities = []

    year = 2024  # Adjust the year as needed

    # Loop over each athlete
    for idx, row in df.iterrows():
        athlete_id = str(row['Athlete ID'])
        name = row['Competitor']
        print(f"\nProcessing data for athlete: {name} (ID: {athlete_id})")

        # Navigate to the athlete's main page
        athlete_url = f'https://www.strava.com/athletes/{athlete_id}'

        for week_number in range(start_week, end_week + 1):
            week_number_str = f"{week_number:02d}"
            week_id = f"interval-{year}{week_number_str}"
            print(f"\nProcessing Week ID: {week_id}")

            # Navigate to the athlete's main page each time
            driver.get(athlete_url)
            time.sleep(1)

            try:
                # Find the week element
                week_element = driver.find_element(By.ID, week_id)

                # Click on the week link
                link_element = week_element.find_element(By.TAG_NAME, 'a')
                link_element.click()
                time.sleep(5)

                # Extract weekly metadata as before
                # Wait for date range element
                date_range_element = WebDriverWait(driver, 15).until(
                    EC.presence_of_element_located((By.ID, "interval-value"))
                )
                date_range = date_range_element.text

                # Extract totals
                totals_section = driver.find_element(By.ID, "totals")
                total_values = totals_section.find_elements(By.TAG_NAME, "strong")

                # Extract distance
                distance_raw = total_values[0].text.strip()
                match = re.match(r'([\d,\.]+)\s*([a-zA-Z]+)', distance_raw)
                if match:
                    distance_value_str = match.group(1).replace(',', '')
                    unit = match.group(2)
                    distance_value = float(distance_value_str)
                else:
                    distance_value = 0.0
                    unit = None

                if unit == 'mi':
                    distance_km = distance_value * 1.60934
                elif unit == 'km':
                    distance_km = distance_value
                else:
                    distance_km = 0.0

                # Extract time
                time_value = total_values[1].text.strip()

                # Extract elevation
                elevation_raw = total_values[2].text.strip()
                match_elevation = re.match(r'([\d,\.]+)\s*([a-zA-Z]+)', elevation_raw)
                if match_elevation:
                    elevation_value_str = match_elevation.group(1).replace(',', '')
                    elevation_unit = match_elevation.group(2)
                    elevation_value = float(elevation_value_str)
                else:
                    elevation_value = 0.0
                    elevation_unit = None

                if elevation_unit == 'ft':
                    elevation_meters = elevation_value * 0.3048
                elif elevation_unit == 'm':
                    elevation_meters = elevation_value
                else:
                    elevation_meters = 0.0

                # Append data to lists
                athlete_ids.append(athlete_id)
                names.append(name)
                week_numbers.append(week_number_str)
                date_ranges.append(date_range)
                distances.append(distance_km)
                times_list.append(time_value)
                elevations.append(elevation_meters)

                print(f"Week {week_number_str}: Distance {distance_km} km, Time {time_value}, Elevation {elevation_meters} m")

                # Extract individual activity data from data-react-props
                # Find the element containing data-react-props
                data_react_props_element = driver.find_element(By.CSS_SELECTOR, '[data-react-props]')
                data_react_props = data_react_props_element.get_attribute('data-react-props')

                # Preprocess the JSON string
                clean_json_str = preprocess_json_string(data_react_props)

                # Parse the JSON
                try:
                    json_data = json.loads(clean_json_str)
                except json.JSONDecodeError as e:
                    print(f"JSON decoding failed: {e}")
                    continue  # Skip to the next week

                # Extract individual activities
                pre_fetched_entries = json_data.get('appContext', {}).get('preFetchedEntries', [])
                for entry in pre_fetched_entries:
                    activity_data = entry.get('activity', {})
                    athlete_data_json = activity_data.get('athlete', {})

                    # Extract stats
                    stats_list = activity_data.get('stats', [])
                    stats = {stat['key']: stat['value'] for stat in stats_list}

                    # Build the activity dictionary
                    activity_info = {
                        'Athlete ID': athlete_data_json.get('athleteId'),
                        'Athlete Name': athlete_data_json.get('athleteName'),
                        'Activity ID': activity_data.get('id'),
                        'Activity Name': activity_data.get('activityName'),
                        'Description': clean_html(activity_data.get('description')),
                        'Start Date': activity_data.get('startDate'),
                        'Elapsed Time': activity_data.get('elapsedTime'),
                        'Type': activity_data.get('type'),
                        'Location': activity_data.get('timeAndLocation', {}).get('location'),
                        'Stat One': stats.get('stat_one'),
                        'Stat Two': stats.get('stat_two'),
                        'Stat Three': stats.get('stat_three'),
                        'Stat One Subtitle': stats.get('stat_one_subtitle'),
                        'Stat Two Subtitle': stats.get('stat_two_subtitle'),
                        'Stat Three Subtitle': stats.get('stat_three_subtitle'),
                        'Week Number': week_number_str
                    }
                    activities.append(activity_info)

                # Wait before the next iteration
                time.sleep(1)

            except Exception as e:
                # Handle cases where there are no workouts or elements are not found
                print(f"Week {week_number_str}: No data available or an error occurred: {e}")
                # Optionally, append zero or None values for weekly data
                athlete_ids.append(athlete_id)
                names.append(name)
                week_numbers.append(week_number_str)
                date_ranges.append(f"Week {week_number_str} - No Data")
                distances.append(0.0)
                times_list.append("0h 0m")
                elevations.append(0.0)
                continue  # Proceed to the next week

    # Create DataFrame with the collected weekly data
    df_weekly = pd.DataFrame({
        'Athlete ID': athlete_ids,
        'Name': names,
        'Week Number': week_numbers,
        'Date Range': date_ranges,
        'Distance (km)': distances,
        'Time': times_list,
        'Elevation (m)': elevations
    })

    # Create DataFrame with individual activities
    df_activities = pd.DataFrame(activities)

    # Further processing for activities DataFrame
    # Extract and clean stats
    df_activities['Distance (km)'] = df_activities['Stat One'].apply(extract_numeric)
    df_activities['Pace (min/km)'] = df_activities['Stat Two'].apply(extract_numeric)
    df_activities['Time'] = df_activities['Stat Three'].apply(clean_html)

    # Convert 'Start Date' to datetime
    df_activities['Start Date'] = pd.to_datetime(df_activities['Start Date'], errors='coerce')

    # Drop the original stat columns
    df_activities.drop(['Stat One', 'Stat Two', 'Stat Three',
                        'Stat One Subtitle', 'Stat Two Subtitle', 'Stat Three Subtitle'], axis=1, inplace=True)

    return df_weekly, df_activities
```

```python
# Extract distance
distance_raw = total_values[0].text.strip()
match = re.match(r'([\d,\.]+)\s*([a-zA-Z]+)', distance_raw)
if match:
    distance_value_str = match.group(1).replace(',', '')
    unit = match.group(2)
    distance_value = float(distance_value_str)
    print(f"Extracted Distance: {distance_value} {unit}")
else:
    print(f"Could not parse distance: {distance_raw}")
    distance_value = None
    unit = None

# Convert distance to kilometers if necessary
if unit == 'mi':
    distance_km = distance_value * 1.60934
elif unit == 'km':
    distance_km = distance_value
else:
    distance_km = None

print(f"Distance in Kilometers: {distance_km} km")
```

```python
# Extract time
time_value = total_values[1].text.strip()
print(f"Time: {time_value}")
```

```python
# Extract elevation
elevation_raw = total_values[2].text.strip()
match_elevation = re.match(r'([\d,\.]+)\s*([a-zA-Z]+)', elevation_raw)
if match_elevation:
    elevation_value_str = match_elevation.group(1).replace(',', '')
    elevation_unit = match_elevation.group(2)
    elevation_value = float(elevation_value_str)
    print(f"Extracted Elevation: {elevation_value} {elevation_unit}")
else:
    print(f"Could not parse elevation: {elevation_raw}")
    elevation_value = None
    elevation_unit = None

# Convert elevation to meters if necessary
if elevation_unit == 'ft':
    elevation_meters = elevation_value * 0.3048
elif elevation_unit == 'm':
    elevation_meters = elevation_value
else:
    elevation_meters = None

print(f"Elevation in Meters: {elevation_meters} m")
```
